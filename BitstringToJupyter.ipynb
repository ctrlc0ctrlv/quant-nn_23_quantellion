```python
import pandas as pd
```

```python
#lecturer, group, subject, day, time
def lsgdt(index):
      
    time = index % 8
    index//=8
    
    day = index % 12
    index//= 12
    
    subject = index % 2
    index//=2
    
    group = index % 2
    index//=2
    
    return (index,group,subject,day,time)
```

```python
lecturer_dict = {0: 'Иванов', 1: 'Петров', 2: 'Сидоров', 3: 'Карпов', 4: 'Соколов'}
days_dict = {0:'пн', 1:'вт', 2:'ср', 3:'чт', 4:'пт', 5:'сб', 6:'вс'}
hours_dict = {0: '9:00-10:00', 1 :'10:00-11:00', 2:'11:00-12:00', 3: '12:00-13:00', 4:'13:00-14:00', 5: '14:00-15:00', 6: '15:00-16:00', 7:'16:00-17:00'}
courses_dict = {0: ('Квантовая механика', 'Квантовая теория информации'),
                1: ('Квантовые вычисления', 'Сложность квантовых алгоритмов'),
                2: ('Квантовые алгоритмы в логистике', 'Квантовое машинное обучение'), 
                3: ('Моделирование квантовых систем', 'Квантовые алгоритмы в химии'),
                4: ('Физическая реализация квантовых компьютеров','Моделирование квантовых алгоритмов')}
groups_dict = {0: 'QC_1', 1: 'QC_2'}
```

```python
def bistring_to_dataframe(bitstring): #bitstring is iterable, consists of 0 and 1
    
    rows_list = []
    for index in range(len(bitstring)):
        if (bitstring[index] == '1'):
            lecturer,group,subject,day,time = lsgdt(index)
            week = day//6
            day = day%6
            
            timetable_string = {'Group' : groups_dict[group],
                                'Weel' : 'неделя ' + str(week),
                                'Day': days_dict[day],
                                'Hour': hours_dict[time],
                                'Subject': courses_dict[lecturer][subject],
                                'Name': lecturer_dict[lecturer],
                               }
            rows_list.append(timetable_string)
            
            
    return pd.DataFrame(rows_list)
```
